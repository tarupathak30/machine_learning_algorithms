```python
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt

x = np.array([[1, 2], [2, 2], [2, 3], [8, 7], [25, 80]])
```

```python
plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

```python
db = DBSCAN(eps=10, min_samples=2)

db.fit(x)
db.labels_

# that's how it works for outliers
```

```python
from sklearn.datasets import make_circles

# creating a concentric circle dataset
x, _ = make_circles(n_samples=500, factor=0.5, noise=0.03, random_state=4)

# applying DBSCAN to the dataset
dbscan = DBSCAN(eps=0.2, min_samples=5)
clusters=dbscan.fit_predict(x)
```

```python
plt.scatter(x[:, 0], x[:, 1], cmap='viridis')
```

```python
plt.scatter(x[:, 0], x[:, 1], c=clusters, cmap='viridis')
```
