```python
from sklearn.datasets import make_classification
import numpy as np
x, y = make_classification(n_samples=100, n_features=2, n_informative=2, n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41, hypercube=False, class_sep=12)
```

```python
import matplotlib.pyplot as plt
```

```python
def step(x):
  return 1 if x>= 0 else 0
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='winter', s=100)
```

```python
x.shape
```

```python
def perceptron(x, y):
  x = np.insert(x, 0, 1, axis=1)
  weights = np.ones(x.shape[1])
  learning_rate = 0.1

  for i in range(1000):
    j = np.random.randint(0, 100) #randomly selecting one data point per iteration(called stochastic update)
    y_hat = step(np.dot(x[j], weights))
    weights = weights + learning_rate * (y[j] - y_hat) * x[j]


  return weights
```

```python
weights = perceptron(x, y)
```

```python
x_input = np.linspace(-3, 3,  100) #create the x values
y_input = -(weights[0]/weights[2] + (weights[1]*x_input)/weights[2])
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='winter', s=100)
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.ylim(-3, 2)
```

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(x, y)
```

```python
m = -(lor.coef_[0][0] / lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3, 3, 100)
y_input1 = m * x_input + b
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x_input, y_input, color='red', linewidth=3)
plt.plot(x_input1, y_input1, color='black', linewidth=3)
plt.scatter(x[:, 0], x[:, 1], c=y, cmap='winter', s=100)
plt.ylim(-3, 2)
```
